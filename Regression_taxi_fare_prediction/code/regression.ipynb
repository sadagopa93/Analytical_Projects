```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

```python
taxi = pd.read_csv(r"D:\project\Analytical_Projects\Regression_taxi_fare_prediction\datasets\taxi.csv")
fares = pd.read_csv(r"D:\project\Analytical_Projects\Regression_taxi_fare_prediction\datasets\fares.csv")
trips = pd.read_csv(r"D:\project\Analytical_Projects\Regression_taxi_fare_prediction\datasets\trips.csv")
```

```python
df1 = taxi.merge(fares,on="medallion",how="left")
```

```python
df1.info()
```

```python
data = df1.merge(trips,on="medallion",how="left")
data.info()
```

```python
data = data.drop_duplicates()
data = data.dropna(subset="medallion")
data.info()
```

```python
data.columns
```

```python
# remove unwanted columns.

df_cols = ["pickup_datetime_y","tip_amount_y","fare_amount_y","hack_license_y","vendor_id_y","trip_time_in_secs_y","pickup_longitude_y","pickup_latitude_y"]
data = data.drop(columns=df_cols)
data.columns
```

```python
# cloumn renameing to make the data_clean

data = data.rename(columns={'pickup_latitude_x' : 'pickup_latitude'})

data.columns
```

```python
# cloumn renaming 

data = data.rename(columns={'pickup_datetime_x' : 'pickup_datetime', 'pickup_longitude_x' : 'pickup_longitude', 'trip_time_in_secs_x' : 'trip_time_in_secs', 'fare_amount_x': 'fare_amount', 'tip_amount_x' : 'tip_amount', 'hack_license_x': 'hack_license', 'vendor_id_x': 'vendor_id'})

data.columns

```

```python
data = data.drop_duplicates()
data.info()
```

```python
data.describe()
```

```python
# check core relation.

data.corr(numeric_only=True)
```

```python
data = data.head(20000)

y = data[['fare_amount']]
y.head()
```

```python
X = data[['pickup_latitude','pickup_longitude','dropoff_latitude','dropoff_longitude','trip_time_in_secs','trip_distance','passenger_count']]
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# Linear Regression 

model_1 = LinearRegression() # Linear regression model
model_2 = Lasso() #Lasso
model_3 = DecisionTreeRegressor() #Decision Tree
model_4 = RandomForestRegressor() #Random Forst
model_5 = SVR() # SVM Regressor
```

```python
model_1.fit(X_train, y_train)
model_2.fit(X_train, y_train)
model_3.fit(X_train, y_train)
model_4.fit(X_train, y_train)
model_5.fit(X_train, y_train)
```

```python
y_pred_1 = model_1.predict(X_test) # Linear regression
y_pred_2 = model_2.predict(X_test) # Lasso regression
y_pred_3 = model_3.predict(X_test) # Descission Regression
y_pred_4 = model_4.predict(X_test) # Random froest
y_pred_5 = model_5.predict(X_test) # SVM
```

```python
# Accuracy for model_1 : Linear Regression

MAE= metrics.mean_absolute_error(y_test, y_pred_1)
MSE= metrics.mean_squared_error(y_test, y_pred_1)
RMSE= np.sqrt(MSE)
R_Squared = metrics.r2_score(y_test,y_pred_1)

print(MAE) # Mean difference error
print(MSE) # Mean Difference Square
print(RMSE) # Mean Square root error
print(R_Squared) # Model accuracy
```

```python
#Lasso
MAE= metrics.mean_absolute_error(y_test, y_pred_2)
MSE= metrics.mean_squared_error(y_test, y_pred_2)
RMSE= np.sqrt(MSE)
R_Squared = metrics.r2_score(y_test,y_pred_2)

print(MAE) # Mean difference error
print(MSE) # Mean Difference Square
print(RMSE) # Mean Square root error
print(R_Squared) # Model accuracy
```

```python
#Descission Tree

MAE= metrics.mean_absolute_error(y_test, y_pred_3)
MSE= metrics.mean_squared_error(y_test, y_pred_3)
RMSE= np.sqrt(MSE)
R_Squared = metrics.r2_score(y_test,y_pred_3)

print(MAE) # Mean difference error
print(MSE) # Mean Difference Square
print(RMSE) # Mean Square root error
print(R_Squared) # Model accuracy
```

```python
#Random Forest:

MAE= metrics.mean_absolute_error(y_test, y_pred_4)
MSE= metrics.mean_squared_error(y_test, y_pred_4)
RMSE= np.sqrt(MSE)
R_Squared = metrics.r2_score(y_test,y_pred_4)

print(MAE) # Mean difference error
print(MSE) # Mean Difference Square
print(RMSE) # Mean Square root error
print(R_Squared) # Model accuracy
```

```python
#SVM
MAE= metrics.mean_absolute_error(y_test, y_pred_5)
MSE= metrics.mean_squared_error(y_test, y_pred_5)
RMSE= np.sqrt(MSE)
R_Squared = metrics.r2_score(y_test,y_pred_5)

print(MAE) # Mean difference error
print(MSE) # Mean Difference Square
print(RMSE) # Mean Square root error
print(R_Squared) # Model accuracy
```
